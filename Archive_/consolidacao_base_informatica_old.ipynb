```python
# Ordenar os contratos pela coluna IDCONTRATOCH e pela DATA_SOURCE_NEW
df_ch = df_ch.sort_values(by=['IDCONTRATOCH', 'DATA_SOURCE_NEW'])

# Criar uma coluna auxiliar para armazenar o tipo de taxa de juro do mês anterior por contrato
df_ch['TIPO_TAXA_ANTERIOR'] = df_ch.groupby('IDCONTRATOCH')['IDTIPOTAXAJURO'].shift()

# Preencher os valores NaN na coluna 'TIPO_TAXA_ANTERIOR' com o mesmo tipo da primeira taxa encontrada para cada contrato
df_ch['TIPO_TAXA_ANTERIOR'] = df_ch.groupby('IDCONTRATOCH')['TIPO_TAXA_ANTERIOR'].fillna(method='bfill')

# Criar uma coluna 'CONTAGEM' para contar o número de ocorrências de cada contrato
df_ch['CONTAGEM'] = df_ch.groupby('IDCONTRATOCH')['IDCONTRATOCH'].transform('count')

# Criar uma coluna 'ALTERACAO_TX' e inicializar com 'Não'
df_ch['ALTERACAO_TX'] = 'Não'

# Identificar os índices onde ocorre alteração de taxa de juro
change_indices = df_ch[df_ch['IDTIPOTAXAJURO'] != df_ch['TIPO_TAXA_ANTERIOR']].index

# Preencher 'Sim' nos casos onde houve mudança de taxa de juro
df_ch.loc[change_indices, 'ALTERACAO_TX'] = 'Sim'

# Identificar os índices onde o contrato aparece apenas uma vez e ajustar 'ALTERACAO_TX' para 'Não'
single_occurrence_indices = df_ch[df_ch['CONTAGEM'] == 1].index
df_ch.loc[single_occurrence_indices, 'ALTERACAO_TX'] = 'Não'

# Excluir colunas auxiliares
df_ch = df_ch.drop(['TIPO_TAXA_ANTERIOR', 'CONTAGEM'], axis=1)

# Exibir o DataFrame
print(df_ch)
```

```python
# Condições para atribuir valores à nova coluna com base na coluna 'PRAZOCONTRATADOANOS'
conditions = [
    (df_ch['PRAZOCONTRATADOANOS'] <= 25),
    (df_ch['PRAZOCONTRATADOANOS'] <= 30),
    (df_ch['PRAZOCONTRATADOANOS'] <= 35),
    (df_ch['PRAZOCONTRATADOANOS'] <= 100)
]

# Valores correspondentes a cada condição
values = [25, 30, 35, 40]

# Aplicar as condições e atribuir os valores à nova coluna 'NOVA_COLUNA'
df_ch['MATURIDADE BUCKETS'] = np.select(conditions, values, default=0)
```

```python
# Aplicar a lógica condicional para criar a nova coluna 'Euribor'
df_ch['EURIBOR'] = np.where(df_ch['TAXA'] == 'Variável', df_ch['TANATUAL'] - df_ch['SPREADATUAL'], 0)
```
